```python
import pandas as pd
import datetime
df = pd.read_csv("data/netflix_titles.csv")
cols = list(df.columns)
```

Look at what and how much data is missing

```python
print(df.isnull().mean())
```

Removing all the trailing spaces from strings

```python
str_cols = list(df.columns)
str_cols.remove('release_year')

for i in str_cols :
    df[i] = df[i].str.strip()
    
```

Replacing all the missing values with "".

```python
columns = ['director', 'cast', 'country', 'rating', 'date_added']
for i in columns :
    df[i] = df[i].fillna("")
```

Working with mixed values 
-> Dates 
    1. Storing the indices with empty values in emp
    2. If index is in emp replace it with 0 Else separate it in month and year
    3. Convert the month string to number, January -> 1, September -> 9

```python
emp = []
for i in range(len(df)) :
    if df['date_added'].iloc[i] == "":
        emp.append(i)
```

```python
month = []
year = []
for i in range(len(df)) :
    if i in emp :
        month.append(0)
        year.append(0)
    else :
        date = df['date_added'].iloc[i].split(" ")
        month.append(date[0])
        year.append(int(date[2]))
        
print(month)
```

```python
for i, val in enumerate(month) :
    if val != 0 :
        dt = datetime.strptime(val, "%B")
        month[i] = dt.month
        
print(set(month))
        
```
